```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
def bias_analysis(symbol):
    history_df = pd.read_csv(f"D:/trading/quant/data/futures/{symbol}.csv")
    history_df['ma20'] = history_df['close'].rolling(20).mean()
    history_df['ma50'] = history_df['close'].rolling(50).mean()
    plt.plot(history_df["close"])
    plt.plot(history_df["ma20"])
    plt.plot(history_df["ma50"])
    plt.show()
    
    bias = (history_df["close"] - history_df["ma20"]) / history_df["ma20"] * 100
    bias = bias[200:]
    print(np.mean(bias), np.std(bias), np.quantile(sorted(bias[100:]), [0.1, 0.25, 0.5, 0.75, 0.9]))
    plt.plot(bias)
    plt.show()

    bias = (history_df["close"] - history_df["ma50"]) / history_df["ma50"] * 100
    bias = bias[200:]
    print(np.mean(bias), np.std(bias), np.quantile(sorted(bias[100:]), [0.1, 0.25, 0.5, 0.75, 0.9]))
    plt.plot(bias)
    plt.show()
```

```python
bias_analysis('ma0')
```

```python
bias_analysis('v0')
```

```python
bias_analysis('rb0')
```

```python
bias_analysis("pp0")
```
